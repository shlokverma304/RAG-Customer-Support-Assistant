## **LOAD PDF**

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = PyPDFLoader("support.pdf")
docs = loader.load()

print(docs[0].page_content[:300])
```

## **CHUNKING**

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_documents(docs)

print("Total chunks:", len(chunks))
```

## **SIMPLE RETRIEVER**

```python
def retrieve(query):
    results = []
    for chunk in chunks:
        if any(word in chunk.page_content.lower() for word in query.lower().split()):
            results.append(chunk.page_content)
    return results
```

## **RAG PIPELINE + HITL**

```python
def rag_system(query):
    docs = retrieve(query)

    if not docs:
        return "Escalate to human agent (HITL)"

    return "Answer:\n" + docs[0][:200] + "..."
```

## **TEST**

```python
print(rag_system("What is refund policy?"))
print(rag_system("How to contact support?"))
print(rag_system("random unrelated question"))
```

## Explanation

- PDF is loaded and split into chunks  
- Query is matched with relevant chunks  
- If no match found → HITL triggered  
- This simulates a RAG pipeline
